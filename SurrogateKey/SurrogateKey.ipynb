### Create DataFrame

# Natural Key vs Surrogate Key 
# - Demo

```python
customers = spark.createDataFrame([
    (1,"Bob","Bob@gmail.com"),
    (2,"Mel","Mel@gmail.com")
],["CustomerID","Name","Email"])

orders = spark.createDataFrame([
    (101,"Bob@gmail.com","Laptop"),
    (130,"Mel@gmail.com","Keyboard")
],["OrderID","Email","Product"])

```

# Customers and Orders 
# Tables

```python
customers.show()

orders.show()
```

### JOIN Orders and Customers on Email
### as a NATURAL KEY

```python
orders.join(customers,"Email").show()
```

### Update in table: Customers
#### Bob's Email changed

```python
customers = spark.createDataFrame([
    (1,"Bob","Robert@gmail.com"),
    (2,"Mel","Mel@gmail.com")
],["CustomerID","Name","Email"])

customers.show()
```

### Join again on Email

```python
orders.join(customers,"Email").show()
```

### Create the table with SURROGATE KEY
#### Add Column : "CustomerID" into ORDERS table

```python
orders = spark.createDataFrame([
    (101, 1, "Laptop"),
    (130, 2,"Keyboard")
], ["OrderID", "CustomerID", "Product"])

orders.show()
```

### JOIN Orders and Customers on CustomerID
### as a SURROGATE KEY 

```python
orders.join(customers, "CustomerID").show()
```
